1.Q1) Given a list of 2-D coordinates, write a function that returns True
andard libraries.

 collinear([[1,1], [2,2], [4,4], [-10, -10]]) = True
 collinear([[1,0], [2,0], [3,1]]) = False

```python
def collinear(points):
    if len(points) < 3:
        return True

    x0, y0 = points[0]
    x1, y1 = points[1]

    for i in range(2, len(points)):
        x, y = points[i]
        if (x1 - x0) * (y - y0) != (x - x0) * (y1 - y0):
            return False
    return True

# Test cases
print(collinear([[1,1], [2,2], [4,4], [-10, -10]]))  # True
print(collinear([[1,0], [2,0], [3,1]]))              # False
```

Q2) Define a function that returns the frequencies of the last digits
   of a list of nonnegative integers.
Use only the Python standard libraries.
Given the list [49, 10, 20, 5, 30, 785]:
9 is the last digit once (in 49),
0 is the last digit three times (in 10, 20 and 30)
5 is the last digit two times (in 785)
last_digit_counts([49, 10, 20, 5, 30, 785])
 = {9:1, 0:3, 5:2} # or something equivalent


```python
def last_digit_counts(nums):
    counts = {}
    for num in nums:
        last_digit = num % 10
        counts[last_digit] = counts.get(last_digit, 0) + 1
    return counts

# Test case
print(last_digit_counts([49, 10, 20, 5, 30, 785]))  # {9: 1, 0: 3, 5: 2}
```
