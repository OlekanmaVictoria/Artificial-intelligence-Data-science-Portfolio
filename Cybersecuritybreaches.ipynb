```python
Cols = ['Name_of_Covered_Entity','State','Date_Of_Breach,Type_Of_Breach','Year','Individuals_Affected']
```

```python
year_Order = df['year'].value_counts().index
state_order = df['State'].value_counts().index
attack_order = df['Type_of_Breach'].value_counts().index
```

EXPLORATION WITH A VISUALIZATION
1)Which Year has the highest number of Attacks.
2)What is the Most common Cyber attack
3)What is the most reported Cyber attack State.

```python
base_color  =   sb.color_palette()[0]
```

```python
plt.figure(figsize=(20,8))
sb.countplot(data=df,y='year',color=base_color,order = year_Order)
plt.xticks (rotation=50)
```

From our Visualization we can see that the year with the highest attacks is 2013.

```python
plt.figure(figsize=(10,6))
sb.countplot(data=df,y = 'Type_of_Breach',color = base_color,order=attack_order)
```

The most prominent attack tend to be Theft,followed by unauthorized access/Disclosure.

```python
plt.figure(figsize=(20,5))
sb.countplot(data=df,x = 'State',color=base_color,order = state_order)
plt.xticks(rotation = 45)
```

The cites with the most Cyber attacks in order of Sequence.
California
Taxes.

Finding Relationships with Visulizations between
Year and individuals attacked
Year and Cyberattacks.

```python
plt.figure(figsize=(20,8))
plt.scatter(data = df, y = 'Individuals_Affected', x = 'year', alpha=1/3)
plt.xticks(rotation=15)
```

```python
plt.hist2d(data = df, y= 'Individuals_Affected', x = 'year',cmin =0.5,cmap = 'viridis_r')
plt.colorbar();
```

There is a positive Relationship between them.

```python
sb.violinplot(data=df, y='Type_of_Breach', x='year',color=base_color,inner=None,order=attack_order)
plt.xticks(rotation=15)
```

```python
sb.boxplot(data=df, y='Type_of_Breach', x='year', color=base_color,order=attack_order)
plt.xticks(rotation=15)
```

```python
ct_counts=df.groupby(['Type_of_Breach','year']).size()
ct_counts=ct_counts.reset_index(name='count')
ct_counts=ct_counts.pivot(index='Type_of_Breach', columns= 'year',values='count')
ct_counts.head()
```

```python
sb.heatmap(ct_counts,annot= True,cmap='vlag_r',center=0)
plt.xticks(rotation=15);
```

Findings.
Attacks increases with time.
Theft was the highest occurance
California is the state with the most cyber attacks
The year with the most Cyber attacks is 2013
The period with most attacks 2009-2013
People affected by Cyber attacks is on the increasing side.
Theft and Hacking IT were the most of the attacks.
