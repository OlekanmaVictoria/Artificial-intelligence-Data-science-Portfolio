```python
# Define dataset directory
dataset_dir = "cats_and_dogs_dataset"
```

# **DATA PREPROCESSING**
I preprocessed the images using the ImageDataGenerator from Keras.
This includes resizing images, data augmentation (such as rotation, shifting, and flipping), and normalization.
We split the dataset into training and testing sets and create data generators for both.


```python
# Define image size and batch size
#the images will be resized to a square shape with both width and height equal to 150 pixels.
#to ensure uniformity,model compartability,memory efficiency and performance.
img_size = (150, 150)
batch_size = 32
```

```python
# Define data generators for training, validation, and testing
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
# Define data augmentation for testing data
test_datagen = ImageDataGenerator(rescale=1.0 / 255)
```

```python
# Define dataset directory
train_dir = os.path.join(dataset_dir, "train")
test_dir = os.path.join(dataset_dir, "test")

# Define data generators for training and testing
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)
```
